```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/My Drive/UFMG/Reconhecimento de padrões/00.Data/heart.csv')
```

```python
data
```

```python
data['class'] = data['class'].map({'absent': -1, 'present': 1})
data
```

```python
def fscorefunc(data):
  class1_data = data[data['class'] == -1].drop(columns=['class'])
  class2_data = data[data['class'] == 1].drop(columns=['class'])

  class1_mean = class1_data.mean()
  class2_mean = class2_data.mean()
  overall_mean = data.drop(columns=['class']).mean()

  class1_std = class1_data.std()
  class2_std = class2_data.std()

  numerator = ((class1_mean - overall_mean) ** 2 + (class2_mean - overall_mean) ** 2)
  denominator = class1_std + class2_std
  fscore = numerator / denominator

  fscore_sorted = fscore.sort_values(ascending=False)

  return fscore_sorted

fscore_result = fscorefunc(data)
print("Fscore Index:")
print(fscore_result)
```
