The objective of this work is to evaluate the implementation of different convolution filters in images


Below are the libraries used to solve this task


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from sklearn.datasets import fetch_olivetti_faces
```

Once we have imported the fetch_olivetti_faces package that contains the dataset that the exercise is based on, we can assign it to a variable as shown below. As can be seen below, the information for each image is represented as a vector of numbers ranging from 0 to 1 , indicating the intensity of the color in each pixel


```python
faces = fetch_olivetti_faces(shuffle=True, random_state=42).data
indice_imagem = 0
imagem_original = faces[indice_imagem, :]
imagem_original
```

A partir do momento que a base de dados ja foi importada e devidamente manipulada, pode-se então declarar a função disponibilizada no enunciado do exercício

```python
def mostra_imagem(imagem, titulo):
    plt.imshow(imagem.reshape((64, 64)), cmap='gray')
    plt.title(titulo)
    plt.axis('off')
    plt.show()
```

The original image that we will use to apply the filters can be seen below, it consists of the first image in the database used


```python
mostra_imagem(imagem_original, 'Imagem Original')
```

First, the edge filter will be applied, the code responsible for its application and the result can be seen below:


```python
filtro_bordas = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
imagem_bordas = ndimage.convolve(imagem_original.reshape((64, 64)), filtro_bordas)
mostra_imagem(imagem_bordas, 'Filtro de Bordas')
```

For the vertical line filter:


```python
filtro_linhas_verticais = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])
imagem_linhas_verticais = ndimage.convolve(imagem_original.reshape((64, 64)), filtro_linhas_verticais)
mostra_imagem(imagem_linhas_verticais, 'Filtro de Linhas Verticais')
```

For the case of horizontal lines:


```python
filtro_linhas_horizontais = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])
imagem_linhas_horizontais = ndimage.convolve(imagem_original.reshape((64, 64)), filtro_linhas_horizontais)
mostra_imagem(imagem_linhas_horizontais, 'Filtro de Linhas Horizontais')
```

Finally, in the case of the sharpen filter:


```python
filtro_sharpen = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
imagem_sharpen = ndimage.convolve(imagem_original.reshape((64, 64)), filtro_sharpen)
mostra_imagem(imagem_sharpen, 'Filtro Sharpen')
```
