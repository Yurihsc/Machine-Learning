```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
N = 30
m1 = np.array([2, 2])
m2 = np.array([3.5, 4])
m3 = np.array([1.5, 4])

p1 = np.random.randn(N, 2) * 0.6 + m1
p2 = np.random.randn(N, 2) * 0.6 + m2
p3 = np.random.randn(N, 2) * 0.6 + m3

xall = np.concatenate((p1, p2, p3), axis=0)
```

```python
plt.scatter(xall[:, 0], xall[:, 1], c='red', s=10)
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3)

kmeans.fit(xall)

cluster_labels = kmeans.predict(xall)

plt.scatter(xall[:, 0], xall[:, 1], c=cluster_labels, s=10)
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot with Cluster Labels')
plt.show()
```
